```python
data['MAle_Lit']=(data['Male literate population']/data['Male population'])
print(data['MAle_Lit'])
data['MAle_Lit'].max()    #finding men's literacy rates
```

```python
data['Female_Lit']=(data['Female literate population ']/data['Female population'])
print(data['Female_Lit'])
data['Female_Lit'].max()
```

```python
result_fe = data[data['Female_Lit'] == 1]
result_fe=result_fe.iloc[:, [3,5,100]]
result_fe  #sorting highest literacy rates for district and states for females
```

```python
data['TP_Lit']=(data['Female literate population ']+data['Male literate population'])/data['Population']
print(data['TP_Lit'])
data['TP_Lit'].max()
```
